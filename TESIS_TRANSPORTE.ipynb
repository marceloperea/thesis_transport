```python
#Cargar la BD desde mi GDRIVE y crear DATAFRAME (ver si el unicode complicará algo en el futuro)

empresas_todas_df = pd.read_csv ("/content/drive/MyDrive/TESIS/BD/Datos/boleto/Informacion Empresas_todas_marzo2018.csv", sep=";",encoding= 'unicode_escape')

empresas_todas_df.head
```

```python
#Separar hora del campo "HORA"

split_hora = empresas_todas_df["HORA"].str.split('[: ]', expand=True)
split_hora.rename(columns={0:'H',1:'M',2:'S',3:'AM_PM'})
# crear un nuevo df que agrupe el anterior con empresas_todas_df

empresas_todas_df[['H','M','S','AM_PM']]= empresas_todas_df["HORA"].str.split('[: ]', expand=True)
empresas_todas_df

```

```python
# Cambiar nombre de la columna "DIA DE LA SEMANA"

import seaborn as sb

empresas_todas_df.rename(columns={'DIA DE LA SEMANA':'DIA'}, inplace=True)

empresas_todas_df.head()

```

```python
# Corregir campos de la nueva columna "DIA"

empresas_todas_df = empresas_todas_df.replace(['LUNES    '],'LUN')
empresas_todas_df = empresas_todas_df.replace(['MARTES   '],'MAR')
empresas_todas_df = empresas_todas_df.replace(['MIÉRCOLES'],'MIE')
empresas_todas_df = empresas_todas_df.replace(['JUEVES   '],'JUE')
empresas_todas_df = empresas_todas_df.replace(['VIERNES  '],'VIE')
```
